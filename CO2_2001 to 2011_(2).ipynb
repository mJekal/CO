```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import os
import warnings
warnings.filterwarnings("ignore")
plt.style.use('ggplot')
from pandas import Timestamp
```

```python
data = pd.read_csv("Gosan_Flask_CO2_2001-2011.csv")
```

```python
data["Time"] = pd.to_datetime(data["Time"])
```

```python
data
```

```python
plt.figure(figsize=(18,10))
plt.plot(data["Time"], data["CO2"])
plt.xlabel("TIme")
plt.ylabel("CO2 [ppm]")
plt.show()
```

```python
timeSeries = data[["Time","CO2"]]
timeSeries.index = timeSeries.Time
ts = timeSeries.drop("Time",axis=1)
```

```python
ts
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(ts["CO2"], model="additive", period=7)
fig = plt.figure()
fig = result.plot()
fig.set_size_inches(20,15)
```

```python
import statsmodels.api as sm
```

```python
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts, lags=20, ax=ax1)
```

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(ts)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
ts_diff = ts - ts.shift()
plt.figure(figsize=(22,8))
plt.plot(ts_diff)
plt.show()
```

```python
result = adfuller(ts_diff[1:])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
import statsmodels.api as sm

fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(ts_diff[1:], lags=20, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(ts_diff[1:], lags=20, ax=ax2)
```

```python
from pandas import datetime
from statsmodels.tsa.arima.model import ARIMA
data.index.freq = 'M'

model = ARIMA(ts, order=(2,1,2))
model_fit = model.fit()
```

```python
start_index = Timestamp("2002-09-06 17:39")
end_index = Timestamp("2012-02-11 12:30")
forecast = model_fit.predict(start=start_index, end=end_index, typ='levels')
```

```python
plt.figure(figsize=(22,8))
plt.plot(data["Time"], data["CO2"], label="original" )
plt.plot(forecast, label = "predicted")
plt.title("Time Series Forecast")
plt.legend()
plt.show()
```

```python
resi = np.array(data[data["Time"] >= start_index]["CO2"]) - np.array(forecast)
```

```python
plt.figure(figsize=(22, 8))
plt.plot(data.Time[data.Time >= start_index], resi, label="CO2 Levels")
plt.xlabel("Time")
plt.ylabel("CO2")
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resi, lags=20, ax=ax1)
```

```python
result = adfuller(resi)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
from sklearn import metrics
```

```python
def scoring(y_true, y_pred):
    r2 = round(metrics.r2_score(y_true, y_pred) * 100, 3)
    corr = round(np.corrcoef(y_true, y_pred)[0,1],3)
    mape = round(
        metrics.mean_absolute_percentage_error(y_true, y_pred) * 100, 3)
    rmse = round(metrics.mean_squared_error(y_true, y_pred, squared=False),3)
    
    df = pd.DataFrame({
        'R2': r2,
        'Corr': corr,
        'PMSE': rmse,
        'MAPE': mape
    },
        index=[0])
    return df
```

```python
scoring(np.array(data[data["Time"] >= start_index]["CO2"]), np.array(forecast))
```
